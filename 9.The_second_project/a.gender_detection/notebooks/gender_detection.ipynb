```python
train = pd.read_csv('../data/train_data.csv')
test = pd.read_csv('../data/test_data.csv')
train.head()
```

```python
# do some preprocessing :)
train['age'].value_counts()
```

```python
most_frequent = train['is_business'].value_counts().sort_values(ascending=False).index[0]
train['is_business'].fillna(most_frequent, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
train['gender'] = le.fit_transform(train['gender'])
to_label = dict(zip(le.classes_, le.transform(le.classes_)))
to_class = dict(zip(le.transform(le.classes_), le.classes_))
print(to_label)
train.head()
```

```python
train.info()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import re
from hazm import WordTokenizer

tokenizer = WordTokenizer()

def my_tokenizer(text):
    result = tokenizer.tokenize(text)
    return result

cv = CountVectorizer()

X_train = train.loc[:6000, 'fullname'].values
y_train = train.loc[:6000, 'gender'].values
X_val = train.loc[6001:, 'fullname'].values
y_val = train.loc[6001:, 'gender'].values
X_test = test.loc[:, 'fullname'].values

train_vectors = cv.fit_transform(X_train)
val_vectors = cv.transform(X_val)
test_vectors = cv.transform(X_test)

print(train_vectors.shape, val_vectors.shape)
```

    حال که داده را پاکسازی کرده و احتمالا ویژگی‌هایی را به آن افزوده یا از آن حذف کرده‌اید، وقت آن است که مدلی آموزش دهید که بتواند متغیر هدف این مسئله را پیش‌بینی کند.

```python
# modeling :)
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(train_vectors, y_train)
```

```python
# evaluate model :)
from sklearn.metrics import f1_score, classification_report, confusion_matrix
from  sklearn.metrics  import accuracy_score

predicted = clf.predict(val_vectors)
print(f1_score(y_val,predicted))
```

```python
import zipfile
import joblib

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

submission.to_csv('submission.csv', index=False)
file_names = ['gender_detection.ipynb', 'submission.csv']
compress(file_names)
```
