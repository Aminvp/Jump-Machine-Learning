ابتدا نیاز است فایل مجموعه‌داده (<code>beijing.csv</code>) را بخوانیم.

```python
import pandas as pd
# reading the dataset
df = pd.read_csv('beijing.csv', encoding='gbk', low_memory=False)
```

```python
df
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
کشف داده‌های پرت
</font>
</h2>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
    برای این کار چارک اول (<code>Q1</code>)، دوم (<code>Q2</code>) و سوم (<code>Q3</code>) ویژگی <i>قیمت متر مربع یک خانه»</i> (ستون <code>price</code>) را محاسبه کرده و سپس آن‌هایی را که کوچکتر از <code>Q1 - 1.5 * IQR</code> و یا بزرگتر از <code>Q3 + 1.5 * IQR</code> هستند به عنوان داده‌ی پرت تشخیص می‌دهیم.
</font>
</p>

```python
# calculating quantiles to detect outliers
Q1, Q2, Q3 = df.price.quantile([0.25, 0.5, 0.75])
IQR = Q3 - Q1
df['price_is_outlier'] = df['price'].apply(lambda p: (p > Q3 + 1.5 * IQR) or (p < Q1 - 1.5 * IQR))
df[['price', 'price_is_outlier']].head(3)
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
مصورسازی
</font>
</h2>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
    اکنون می‌توان به کمک کتابخانه‌ی <code>plotly</code> اقدام به مصورسازی داده‌های پرت کنیم.
</font>
</p>

```python
import plotly.express as px
fig = px.box(df, y='price')
fig.update_traces(marker_color='#FF851B')
fig.update_layout(autosize=False, width=600, height=500, title='Detected Outliers')
fig.show()
```
