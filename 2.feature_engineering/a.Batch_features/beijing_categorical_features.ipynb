ابتدا نیاز است فایل مجموعه‌داده‌ی بدون مقادیر گم‌شده (<code>beijing_no_missing.csv</code>) را بخوانیم.

```python
import pandas as pd
# reading the dataset
df = pd.read_csv('beijing_no_missing.csv', encoding='gbk', low_memory=False)
```

```python
df
```

<h2 dir="rtl" align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
تبدیل ویژگی‌های دسته‌ای
</font>
</h2>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
ابتدا نگاهی به مقادیر ستون <code>renovationCondition</code> بیندازیم:
</font>
</p>

```python
df['renovationCondition'].value_counts()
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
ساده‌ترین راه برای تبدیل ستون‌های اسمی به عددی، استفاده از یک دیکشنری برای تبدیل هر دسته به یک عدد است. در این روش، دسته‌ها با استفاده از یک دیکشنری به اعداد <code>0</code> تا <code>n-1</code> نگاشت می‌شوند. (<code>n</code> تعداد دسته‌ها است). با کد زیر می‌توانیم ستون ‍<code>renovationCondition</code> را به عدد تبدیل کنیم:
</font>
</p>

```python
mapping = {'other': 0,
           'hardcover' : 1,
           'Simplicity' : 2,
           'rough' : 3}

df['Mapping'] = df['renovationCondition'].map(mapping)

df['Mapping'].value_counts()
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
به کمک کلاس <code>LabelEncoder</code> در کتابخانه‌ی <code>scikit-learn</code> نیز می‌توانیم همین کار را به شکل ساده‌تر انجام دهیم:
</font>
</p>

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['LabelEncoding'] = label_encoder.fit_transform(df['renovationCondition'])

df['LabelEncoding'].value_counts()
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
بیایید، ستون <code>renovationCondition</code> را بصورت one-hot کدگذاری کنیم. برای این کار می‌توانیم از تابع <code>get_dummies</code> کتابخانه‌ی <code>pandas</code> کمک بگیریم.
</font>
</p>

```python
dummies = pd.get_dummies(df['renovationCondition'])
dummies.head()
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
اگر بخواهیم حاصل این کدگذاری را به دیتافریم اصلی خود اضافه کنیم می‌توانیم از کد زیر کمک بگیریم:
</font>
</p>

```python
encoded_df = pd.concat([df, dummies], axis=1)
```
