ابتدا نیاز است فایل مجموعه‌داده‌ی بدون مقادیر گم‌شده (<code>beijing_no_missing.csv</code>) را بخوانیم.

```python
import pandas as pd
# reading the dataset
df = pd.read_csv('beijing_no_missing.csv', encoding='gbk', low_memory=False)
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
تعریف ویژگی‌های جدید
</font>
</h2>

```python
import datetime
import numpy as np

# creating the featuers
current_year = datetime.datetime.now().year
print('Current year:', current_year)

# Latitude and Longitude of Bejing (from http://latlong.net/place/beijing-china-27567.html)
bejing_center_Lng = 116.383331
bejing_center_Lat = 39.916668

df['number_of_rooms'] = df['livingRoom'] + df['drawingRoom']

# convert the constructionTime column to int
df['constructionTime'] = df['constructionTime'].astype(int)
# calculate the age of the house
df['age'] = current_year - df['constructionTime']

# calculate the Euclidean distance from the city center

df['distance_to_city_center'] = ((df['Lat'] - bejing_center_Lat)**2 + (df['Lng'] - bejing_center_Lng)**2) ** (1/2)

# calculate the angle in degrees
df['angle_degrees'] = np.degrees(np.arctan((df['Lat'] - bejing_center_Lat)/(df['Lng'] - bejing_center_Lng)))

df[['id', 'number_of_rooms', 'age', 'distance_to_city_center', 'angle_degrees']].head()
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
​​همانطور که گفته شد برای گسسته‌سازی دو روش یکنواخت و مبتنی بر چندک وجود دارد. برای این ستون خاص، می‌توان زمان ساخت خانه‌ها را به دهه‌ای که در آن ساخته شده‌اند، تبدیل کرد. در نتیجه، از گسسته‌سازی یکنواخت استفاده می‌کنیم. به عنوان مثال خانه‌ای که در سال ۱۹۹۲ ساخته شده به دهه ۱۹۹۰ و خانه‌ای که در سال ۲۰۰۰ ساخته شده به دهه ۲۰۰۰، اختصاص پیدا می‌کند.​​قطعه کد زیر، این گسسته‌سازی را نشان می‌دهد:
</font>
</p>

```python
# discretize constructionTime to constructionDecade
df['constructionDecade'] = df['constructionTime'].apply(lambda y: str(int(10 * np.floor(y / 10))))

df[['id', 'constructionTime', 'constructionDecade']].head()
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
ویژگی‌های چندجمله‌ای
</font>
</h2>

```python
from sklearn.preprocessing import PolynomialFeatures

# choosing some numeric columns
selected_columns = ['Lng', 'Lat', 'ladderRatio']
df1 = df[selected_columns]

# creating the polynomial features object for two-degree polynomial features
pf = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)

# fitting to the features
pf.fit(df1)

# creating polynomial features
poly_feats = pf.transform(df1)

# creating a dataframe with all the features
num_feats = poly_feats.shape[1]
df_transformed = pd.DataFrame(poly_feats, columns=[f'f_{i}' for i in range(1, num_feats + 1)])
df_transformed.columns = pf.get_feature_names_out()
df_transformed.head()
```
